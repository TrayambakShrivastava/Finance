```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock = ['HDFC' , 'ICICI' , 'ITC']
mu_list = [ 0.000412 , 0.000407 , -0.000220 ]
stdev_list = [ 0.0327 , 0.0354 , 0.0423 ]
drift = mu_list[0]*0.37 + mu_list[1]*0.329 + mu_list[2]*0.3
stdev = math.sqrt((stdev_list[0]**2)*(0.37**2) + (stdev_list[1]**2)*(0.329**2) + (stdev_list[2]**2)*(0.30**2))
```

```python
#drift
mu = drift
#no of steps FOR 252 working days in a year 
n = 252
#time in years
T = 1
#no of sims
N =10
#stddev
sigma = stdev
#initial stock price = 0.37*hdfc + 0.329*icici + 0.3*itc
So = 0.37*2420 + 0.329*600.5 + 0.3*202.4

paths = []
```

```python
dt = T/n
#St = prices[-1]*np.exp((mu - sigma**2/2)*dt + sigma * np.random.normal(0,np.sqrt(dt),))

for i in range(N):
  prices = [So]
  time = 0

  while(time + dt <= T) :
    prices.append(prices[-1]*np.exp((mu - sigma**2/2)*dt + sigma * np.random.normal(0,np.sqrt(dt))))
    time+=dt

  if T - (time) > 0:
    prices.append(prices[-1]*np.exp((mu - sigma**2/2)*(T-(time)) + sigma * np.random.normal(0,np.sqrt(T - (time)))))

  paths.append(prices)


```

```python
for path in paths:
  plt.plot(path)
```
